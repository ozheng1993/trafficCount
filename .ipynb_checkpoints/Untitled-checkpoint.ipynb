```python
import numpy as np
import cv2
import math
rect = (0,0,0,0)
startPoint = False
endPoint = False
counter=0
line_id=0
saved=False
frame_num=0
import os
import csv
font = cv2.FONT_HERSHEY_SIMPLEX
car_counter=0
carInside=False
carOut=False
carCounted=False
rects=[]
carCounteds=[]
carInsides=[]
carCounters=[]

fileName="DJI_0002.MOV"

filePath='data/'+fileName+'.csv'


from sys import argv
csvKeys=["boxId",       
        'frame number', 
        'total', 
    ]
if os.path.exists('data'):
    print("file data exist")
else:
    print("file data not exist,created")
    os.mkdir('data')
if os.path.exists(filePath):
    print("csv exist")
else:
    with open(filePath, 'wb') as outcsv:
        writer = csv.DictWriter(outcsv, fieldnames = csvKeys)
        writer.writeheader()
    outcsv.close()
    print("csv not exist,created")
fgbg = cv2.bgsegm.createBackgroundSubtractorMOG()
def countCar(bx1,by1,bx2,by2,cx,cy):
    if(cx<max(bx1,bx2) and cx>min(bx1,bx2) and cy<max(by1,by2) and cy>=min(by1,by2)):
        print("")
        return True
    
def on_mouse(event,x,y,flags,params):
    global rect,rects,carCounteds,carInsides,carCounters,startPoint,endPoint,x1,x2,y1,y2,dist,counter,line_id,string,saved
    if event == cv2.EVENT_RBUTTONDOWN:
        rects=rects[:-1]
        carCounteds=carCounteds[:-1]
        carInsides=carInsides[:-1]
        carCounters=carCounters[:-1]

    if event == cv2.EVENT_MBUTTONDOWN:
        if startPoint == True and endPoint == True and saved== False:
            rects.append(rect)
            carCounteds.append(False)
            carInsides.append(False)
            carCounters.append(0)
            saved=True
            
#             print(dist)
#             print(counter)
#             #print(dist1)
#             counter+=1
#             print("right clicked")
#             string="saved"
#             csvValuse=[line_id, frame_num,dist]
#             line=str(line_id)+","+str(frame_num)+","+str(dist)+"\n"
#             print(line)
#             with open('data/Raw.csv', 'a') as outcsv:
                
#                 outcsv.write(line)
                
#             outcsv.close()
#             print("Raw.csv insert")
    # get mouse click
    if event == cv2.EVENT_LBUTTONDOWN:

        if startPoint == True and endPoint == True:
            startPoint = False
            endPoint = False
            print(counter)
            print("reset")
            line_id=counter
            saved = False
            rect = (0, 0, 0, 0)

        if startPoint == False:
            rect = (x, y, 0, 0)
            startPoint = True
#             print(x1)
#             print(y1)
#             print(x2)
#             print(y2)
#             print("setpoint1")
        elif endPoint == False:
            rect = (rect[0], rect[1], x, y)
            
#             x1=x1
#             y1=y1
#             x2=x
#             y2=y
            endPoint = True
            #dist1 = math.sqrt( (x2 - x1)**2 + (y2 - y1)**2 )
            #dist = math.hypot(x2 - x1, y2 - y1)
            #string="id: "+str(line_id)+"| length :"+str(dist)+"|"+str(frame_num)
            
            
#             print(x1)
#             print(y1)
#             print(x2)
#             print(y2)
#             print("setpoint2")

cap = cv2.VideoCapture(fileName)
# (grabbed, frame) = cap.read()
# frame = cv2.resize(frame, (0,0), fx=0.5, fy=0.5)
# r = cv2.selectROI(frame)
     
    # Crop image
#imCrop = frame[int(r[1]):int(r[1]+r[3]), int(r[0]):int(r[0]+r[2])]
waitTime = 50

# #Reading the first frame
# (grabbed, frame) = cap.read()
frame_counter=0
while(cap.isOpened()):
#     events = [i for i in dir(cv2) if 'EVENT' in i]
#     print events
    (grabbed, frame) = cap.read()
#     if(frame_counter<=1):
#         frame_counter+=1;
#         print("skipe")
#         continue;
#     frame_counter=0
    frame_num=cap.get(cv2.CAP_PROP_POS_FRAMES) # retrieves the current frame number
    #print(frame_num)
    print(rects)
    frame = cv2.resize(frame, (0,0), fx=0.5, fy=0.5)
    #frame = frame[int(r[1]):int(r[1]+r[3]), int(r[0]):int(r[0]+r[2])]
    cv2.putText(frame,str(frame_num),(10,10), font, 1,(255,255,255),2,cv2.LINE_AA)
    
#     frame = cv2.addWeighted(frame, 3, cv2.blur(frame, (30, 30)), -4, 128)

    
    fgmask = fgbg.apply(frame)
    mask1 = cv2.GaussianBlur(fgmask, (5, 5), 0)
    #mask1 = cv2.adaptiveThreshold(mask1, 120, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 5, 1)
    retval2,mask1 = cv2.threshold(mask1, 150, 255, cv2.THRESH_BINARY)

    #mask1 = cv2.bilateralFilter(mask1,2,5,5)
    mask1 = cv2.morphologyEx(mask1, cv2.MORPH_OPEN, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (4, 4)))
    mask1 = cv2.GaussianBlur(mask1, (9, 9), 0)
    mask1 = cv2.morphologyEx(mask1, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (12, 12)))
    #mask1 = cv2.Canny(mask1, 100, 300)

#     mask1 = cv2.Canny(mask1, 100, 300)
    #mask1 = cv2.morphologyEx(mask1, cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5)))
    #mask1 = cv2.Canny(mask1, 100, 100)

    im2, contours, hierarchy = cv2.findContours(mask1,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    
#     for cnt in contours:
#         hull = cv2.convexHull(cnt)
#         cv2.drawContours(frame, [hull], -1, (0, 0, 255), 3) 
#         for c in [hull]:
#            # calculate moments for each contour
#             M = cv2.moments(c)
#            # calculate x,y coordinate of center
#             if M["m00"] != 0:
#                 cX = int(M["m10"] / M["m00"])
#                 cY = int(M["m01"] / M["m00"])
#             else:
#                 cX, cY = 0, 0
#             for index, item in enumerate(rects):
#                 carInsides[index]=False
#                 if(countCar(item[0], item[1],item[2], item[3],cX,cY)):
#                     cv2.circle(frame, (cX, cY), 5, (0, 255, 0), -1)
#                     cv2.putText(frame, "inside", (cX - 25, cY - 25),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
#                     carInsides[index]=True
#                     carCounteds[index]=True
#                     carCounters[index]+=0
#                     break
#                 else:
#                     cv2.circle(frame, (cX, cY), 5, (0, 0, 255), -1)
#                     cv2.putText(frame, "car", (cX - 25, cY - 25),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
            
            
            
    for index, item in enumerate(rects):
        carInsides[index]=False
        for cnt in contours:
            hull = cv2.convexHull(cnt)
            cv2.drawContours(frame, [hull], -1, (0, 0, 255), 3) 
            for c in [hull]:
               # calculate moments for each contour
                M = cv2.moments(c)
               # calculate x,y coordinate of center
                if M["m00"] != 0:
                    cX = int(M["m10"] / M["m00"])
                    cY = int(M["m01"] / M["m00"])
                else:
                    cX, cY = 0, 0
                if(countCar(item[0], item[1],item[2], item[3],cX,cY)):
                        cv2.circle(frame, (cX, cY), 5, (0, 255, 0), -1)
                        cv2.putText(frame, "inside", (cX - 25, cY - 25),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                        carInsides[index]=True
                        if carCounteds[index]==False:
                            carCounteds[index]=True
                            carCounters[index]+=1
                            line=str(index)+","+str(frame_num)+","+str(carCounters[index])+"\n"
                            print(line)
                            with open(filePath, 'a') as outcsv:

                                outcsv.write(line)

                            outcsv.close()
                            print("Raw.csv insert")
                        break
                else:
                        cv2.circle(frame, (cX, cY), 5, (0, 0, 255), -1)
                        cv2.putText(frame, "car", (cX - 25, cY - 25),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
        if(carInsides[index]==True):
            cv2.putText(frame,"car inside",(item[0]-5, item[1]-20), font, 0.5,(0,255,0),2,cv2.LINE_AA)
        else:
            cv2.putText(frame,"no car inside",(item[0]-5, item[1]-20), font, 0.5,(0,0,255),2,cv2.LINE_AA)
            carCounteds[index]=False



    
    
#     if(countCar(item[0], item[1],item[2], item[3],cX,cY)):
#         cv2.circle(frame, (cX, cY), 5, (0, 255, 0), -1)
#         cv2.putText(frame, "inside", (cX - 25, cY - 25),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
#         carInsides[index]=True
#         carCounteds[index]=True
#         carCounters[index]+=0
#         break
#     else:
#         cv2.circle(frame, (cX, cY), 5, (0, 0, 255), -1)
#         cv2.putText(frame, "car", (cX - 25, cY - 25),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
            
                       

            
            
            
                
     #    for index, item in enumerate(rects):           
                
# #         cv2.putText(frame,"no car indise",(rect[0], rect[1]-10), font, 1,(255,255,255),2,cv2.LINE_AA)
#     if(hasCarInsde==True):
#         cv2.putText(frame,"car inside",(rect[0], rect[1]-20), font, 1,(0,255,0),2,cv2.LINE_AA)
#         if carCounted==False and startPoint == True and endPoint == True:
#             car_counter+=1
#             carCounted=True
#     else:
#         cv2.putText(frame,"no car inside",(rect[0], rect[1]-20), font, 1,(0,0,255),2,cv2.LINE_AA)
#         carCounted=False
        

    
    #cv2.drawContours(frame, contours, -1, (0,255,0), 3)
    cv2.namedWindow('frame')
    cv2.setMouseCallback('frame', on_mouse)    

    #drawing rectangle
    if startPoint == True and endPoint == True:
        cv2.rectangle(frame, (rect[0], rect[1]), (rect[2], rect[3]), (255, 0, 255), 2)
        #cv2.line(frame, (x1, y1), (x2, y2), (255,0,255), 3)
        
    for index, item in enumerate(rects):
        cv2.rectangle(frame, (item[0], item[1]), (item[2], item[3]), (0, 255,0), 2)
        cv2.putText(frame,str(index),(item[0], item[1]-10), font, 0.5,(255,255,255),2,cv2.LINE_AA)
        cv2.putText(frame,str(carCounters[index]),(item[0], item[1]-30), font, 1,(255,0,0),2,cv2.LINE_AA)
        
    cv2.imshow('frame',frame)
    #cv2.imshow('blur',mask)
    key = cv2.waitKey(waitTime) 

    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
